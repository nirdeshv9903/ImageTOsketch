## Create the code that convert the image into a sketch

```python
# First install the opencv-python libary (pip install opencv-python)
# Then import it like that
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# this blok is not nassary
%matplotlib inline
import matplotlib
matplotlib.use("nbAgg")
# Change the default format for the vector formate
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# let's read the image with gray scale
img = cv2.imread('tom1.jpeg', 0)
```

```python
# let's show the image in gray scale
plt.imshow(img, cmap = "gray")
```

```python
img.min()
```

```python
# let's check the maximum value image
img.max()
```

```python
# let's inverse the color of the image
inv = 255-img
```

```python
plt.imshow(inv, cmap = "gray")
```

```python
# let's smothing the image or blur the image 
blur = cv2.GaussianBlur(inv, ksize=(21,21), sigmaX = 0, sigmaY = 0)
```

```python
plt.imshow(blur, cmap = "gray")
```

```python
# implementing dodgeing
dodge = lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
# let's blend's the reale and blur image together
bl = dodge(img, blur)
```

```python
plt.imshow(bl, cmap = "gray")
```

```python
cv2.imwrite('tom_S.jpeg', bl)
```
